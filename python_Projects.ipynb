```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head)
```

```python
df.isnull().sum()
```

**Gender Distribution**

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more tha number of males
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

**Relationship Between Parent's Education and Student Score**

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot= True)
plt.title("Relationship Between Parent's Education and Student Score")
plt.show()
```

```python
#from above chart we hane concludes Parents have good Educaton
```

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1, annot= True)
plt.title("Relationship Between Parent's Maritial Status and Student Score")
plt.show()
```

```python
#from the above chart we concludes there is no/negative impact on the student's
#score due to their Parent's marital status
```

**Distribution of Ethnic Group**

```python
from enum import auto
groupA = df.loc[(df["EthnicGroup"]== "group A")].count()
groupB = df.loc[(df["EthnicGroup"]== "group B")].count()
groupC = df.loc[(df["EthnicGroup"]== "group C")].count()
groupD = df.loc[(df["EthnicGroup"]== "group D")].count()
groupE = df.loc[(df["EthnicGroup"]== "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"],]
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distriution of Groups")
plt.show()
```
